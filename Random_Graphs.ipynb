```python
def get_partial_graph(n, edges, p):
    # Returns a NetworkX graph with all edges
    # that have weight less than p
    
    G = nx.Graph()
    G.add_nodes_from(range(0,n))
    i = 0
    while (edges[i][0] <= p):
        G.add_edge(edges[i][1], edges[i][2])
        i += 1

    return G
```

```python
def get_edges(n):
    edges = []
    for i in range(0,n):
        for j in range(i+1, n):
            edges.append([random.random(), i, j])
    return sorted(edges, key=lambda edge: edge[0])
                
```

```python
get_edges(5)
```

```python
def make_complete_graph(edges):
    G = nx.Graph()
    for edge in edges:
        G.add_edge(edge[1], edge[2], weight=edge[0])
    return G
```

```python
E = get_edges(1000)
```

```python
G = make_complete_graph(E)
```

```python
T = nx.minimum_spanning_tree(G)
```

```python
pos = nx.spring_layout(T)
```

```python
nx.draw(get_partial_graph(1000, E, 0.4), pos)
```

```python
R = lambda p: nx.draw(get_partial_graph(1000, E, p), pos, node_size=2)
```
